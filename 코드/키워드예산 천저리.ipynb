# 추출한 키워드로 전처리 함수 만들기

```python
# 데이터 전처리해주는 함수 생성
def process_data(data, local_code):
    data = data.drop(columns=['번호']) # 쓸모없는 번호열 제거
    data['지출액'] = data['지출액'].str.replace(",", "").str.strip().apply(lambda x: math.inf if x == '' else float(x)) # 지출액 열 형태 변환
    include_keywords = '한부모|미혼모|돌봄|육아|가정|보육'
    exclude_keywords = '철거비용|감량기|협의비용|인공지능|간판|벼룩시장|백세|위촉장|유류비|렌탈|부동산|SOS|세자녀|배달|임금|급량비|연금|상해보험|수수료|LPG사용량|충전비|방문차량|워크숍|수리|기간제근로자|전기레인지|현수막|안전점검|번역|행사비|공연비|출연료|회의|이벤트|강사|간담회|홍보|교체비|케이블|사무관리비|소요경비|증설|다문화|외국인|김장|체육활동|행사운영비|급여|구독료|인쇄|대관료|홍보물품비|반환|간담회비|교사|담임|연장|여비|납부|구입|구매|입양|세차비|소득세|보험료|간식|중식|부서운영|개선비|활동용품|다과|회식|임차료|특근매식비|장애|교육|인건비|육아휴직|반납|요금|직원|수당|공사|활동비|노인|어르신|의료비|제작|공무원'

    data = data[(data['사업개요'].str.contains(include_keywords, regex=True)) &(~data['사업개요'].str.contains(exclude_keywords, regex=True) | data['사업개요'].str.contains('한부모|미혼모', regex=True))| (data['세부사업명'].str.contains('한부모|미혼모', regex=True))]
    lst = data["세부사업명"].unique() # 해당열에 맞는

    processed_data = [] # 리스트 초기화
    for business in lst:
        subset = data[data['세부사업명'] == business].copy() # 해당 세부사업명을 가지고 있는행들을 복사
        total_expense = subset['지출액'].sum() # 지출액 총합 계산
        subset['총 지출액'] = total_expense
        subset = subset.drop(['지급일자', '사업개요', '지출액'], axis=1).iloc[[0]] # 다 같은 행이므로 하나 고르기
        processed_data.append(subset) # 리스트에 추가하기

    combined_data = pd.concat(processed_data) # 리스트에 받은 데이터들 한번에 모아 프레임 형태로 전환
    combined_data['법정동코드'] = local_code # 법정동코드 할당
    return combined_data
```

```python
# 모든 자치구를 자동 전처리시킴

seoul_gu_list = ['서대문구', '마포구', '양천구', '강서구', '구로구', '금천구', '영등포구', '동작구', '서초구', '강남구', '송파구', '강동구',
                 '강북구', '광진구', '노원구', '도봉구', '성동구', '성북구', '용산구', '종로구', '중구', '은평구', '중랑구', '동대문구', '관악구']
processed_data = []

for local_gu in seoul_gu_list:
    local_code = code.loc[(code['서울특별시'] == local_gu), '법정동코드'].iloc[0]
    file_name = f"/content/drive/MyDrive/23 데청캠/전처리/{local_gu}.csv"
    data = pd.read_csv(file_name)
    processed_data.append(process_data(data, local_code))

all_data = pd.concat(processed_data)
```
