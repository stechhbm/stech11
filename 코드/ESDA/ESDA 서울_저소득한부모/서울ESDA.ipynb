```python
# Shapefile 로딩 (경로는 실제 데이터 파일의 경로에 맞게 수정해야 합니다)
gdf = gpd.read_file('/content/drive/MyDrive/데이터청년캠퍼스/공간 데이터 분석/Seoul.shp', encoding = 'utf-8')
```

```python
# 특정 열 선택 (예산 및 한부모가족 수 열을 선택합니다)
budget_column = "TOTAL_BUDG"
count_column = "LOW_INCOME"
facility_column = "WELFARE_FA"
derived_budget = "DERIVED_BU"
budget_values = gdf[budget_column]
count_values = gdf[count_column]
facility_values = gdf[facility_column]
derived_values = gdf[derived_budget]
```

```python
# Queen 가중치 행렬 생성
wq = libpysal.weights.Queen.from_dataframe(gdf)
```

```python
# Spatial Lag 계산
lag_budget = libpysal.weights.lag_spatial(wq, budget_values)
lag_count = libpysal.weights.lag_spatial(wq, count_values)
lag_facility = libpysal.weights.lag_spatial(wq, facility_values)
lag_derived = libpysal.weights.lag_spatial(wq, derived_values)
```

```python
y=lag_count
moran = Moran(y, wq)
moran.I
```

```python
fig, ax = moran_scatterplot(moran, aspect_equal=True)
plt.show
```

```python
m_local=esda.moran.Moran_Local(y,wq)
```

```python
# Plot
fig, ax = moran_scatterplot(m_local, p=0.05)
ax.set_xlabel('derived')
ax.set_ylabel('Spatial Lag of Price')
plt.text(1.95, 0.5, "HH", fontsize=25)
plt.text(1.95, -1.5, "HL", fontsize=25)
plt.text(-2, 1, "LH", fontsize=25)
plt.text(-1, -1, "LL", fontsize=25)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,12))
lisa_cluster(m_local, gdf, p=0.05, figsize = (16,12),ax=ax)
plt.show()
```

```python
# Moran's I 분석
moran_budget = esda.Moran(budget_values, wq)
moran_count = esda.Moran(count_values, wq)
moran_facility = esda.Moran(facility_values, wq)
moran_derived = esda.Moran(derived_values, wq)
```

```python
# Moran's I 결과 출력
print("Moran's I (Count):", moran_count.I)
print("Moran's p-value (Count):", moran_count.p_sim)
```

```python
# Moran's I 결과 출력
print("Moran's I (Budget):", moran_budget.I)
print("Moran's p-value (Budget):", moran_budget.p_sim)
print("Moran's I (Count):", moran_count.I)
print("Moran's p-value (Count):", moran_count.p_sim)
print("Moran's I (facility):", moran_facility.I)
print("Moran's p-value (facility):", moran_facility.p_sim)
print("Moran's I (derived):", moran_derived.I)
print("Moran's p-value (derived):", moran_derived.p_sim)
```

```python
# Moran's I 시각화
plot_moran(moran_budget, zstandard=True, figsize=(10, 4))
plot_moran(moran_count, zstandard=True, figsize=(10, 4))
plot_moran(moran_facility, zstandard=True, figsize=(10, 4))
plot_moran(moran_derived, zstandard=True, figsize=(10, 4))
```

```python
# 공간 상관관계 시각화 (Moran's Scatterplot)
plt.scatter(budget_values, lag_budget, color='blue')
plt.title("Moran's Scatterplot (Budget)")
plt.xlabel(budget_column)
plt.ylabel("Spatial Lag of " + budget_column)
```
