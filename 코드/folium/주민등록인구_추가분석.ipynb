#### json 파일 불러오기

```python
import requests
import json


r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
# json 파일 훑어보기
# 서울 행정구역 json raw파일(githubcontent) -> polygon data, 위도와 경도, 'code'로 구성되어 있음
# 강동구의 코드가 11250 => 확인 결과, 행정동 코드로 판명
seoul_geo
```

### 추가 - 전체 주민등록인구 대비 비율 구하기

```python
import pandas as pd

# str 값으로 가져올 경우 (시군구명) str 변환 불필요
# 총합 시설 알아보기
divData = pd.read_excel('/content/drive/MyDrive/2023 데청캠/프로젝트/한부모수급자_시각화용.xlsx', sheet_name = '중위소득총계DIV')

divData
```

```python
# 데이터 전처리 - 컬럼명 매칭
divData = divData.rename(columns={'지역구':'name'})
divData
```

```python
# folium 패키지로 값 기반 map 생성

import plotly.express as px

# locations : dataframe, geojson 모두 같은 변수명인 값 (code/name)
# color : 시각화 할 때 색으로 드러나는 기준이 되는 값

divFig = px.choropleth(divData, geojson=seoul_geo, locations='name', color='전체_수급가구비',
                                color_continuous_scale='Blues',
                                featureidkey='properties.name')
divFig.update_geos(fitbounds="locations", visible=False)
divFig.update_layout(title_text='서울시 구별 주민등록인구 대비 수급가구', title_font_size=20)
```

```python
# 시각화 파일을 png로 저장
import os

os.chdir('/content/drive/MyDrive/2023 데청캠/프로젝트/divDatas')
divFig.write_image('서울시 구별 주민등록인구 대비 수급가구.png')
# 시각화 파일을 html로 저장 : 클릭해서 내용을 볼 수 있다는 점에서 위 아래 모두 하는 게 좋을 듯
divFig.write_html("서울시 구별 주민등록인구 대비 수급가구.html")
```
